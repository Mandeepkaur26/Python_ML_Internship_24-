```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto = pd.read_csv("Automobile_data.csv")
auto.head()
```

```python
brandVSfuel=auto.groupby(["make","fuel-type"]).size().unstack(fill_value=0)
brandVSfuel
```

```python
brandVSfuel.plot(kind='bar',stacked=True)
plt.show()
```

```python
brandVSdoors=auto.groupby(["make","num-of-doors"]).size().unstack(fill_value=0)
brandVSdoors
```

```python
brandVSdoors.plot(kind='bar',stacked=True)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=auto["wheel-base"],linewidth=1.5,label="Wheel Base")
sns.lineplot(data=auto["engine-size"],linewidth=1.5,label="Engine Size")

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")

plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#F0F00F","axes.grid":False,"xtick.labelsize":10,"ytick.labelsize":10})
plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")
plt.xticks(rotation=45)
sns.lineplot(x=auto["engine-size"],y=auto["wheel-base"].index.values,color='r')

plt.show()

```
